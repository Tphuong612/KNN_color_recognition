```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("./training_dataset/data.csv")
```

```python
# import pandas as pd
# import matplotlib.pyplot as plt

# # Đọc dữ liệu từ file CSV
# df = pd.read_csv('ten_file.csv')

# # Hiển thị dữ liệu đầu tiên để xem cấu trúc
# print(df.head())

# Đếm số lượng mỗi nhãn
label_counts = df['label'].value_counts()

# Vẽ biểu đồ
plt.figure(figsize=(10, 6))
label_counts.plot(kind='bar')
plt.title('Phân phối của nhãn')
plt.xlabel('Nhãn')
plt.ylabel('Số lượng')
plt.xticks(rotation=45)
plt.show()

```
